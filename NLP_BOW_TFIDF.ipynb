# Bag of Words

```python
import re
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
```

```python
ps=PorterStemmer()
lm=WordNetLemmatizer()
sentences3=nltk.sent_tokenize(para)
b=re.sub('[^a-zA-Z]',' ',sentences3[1])
bb=b.split()
```

```python
corpus=[]
for i in range(len(sentences3)):
  review=re.sub('[^a-zA-Z]',' ',sentences3[i])
  review=review.lower()
  review=review.split()
  review=[lm.lemmatize(word) for word in review if word not in set(stopwords.words('english'))]
  review=' '.join(review)
  corpus.append(review)
```

```python
corpus[0:6]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer()
X=cv.fit_transform(corpus).toarray()
```
